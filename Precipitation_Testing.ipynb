```python
###Import the precip df###

#Control#
high = pd.read_csv('CSV_DFs/CF_High_Precip.csv')
low = pd.read_csv('CSV_DFs/CF_Low_Precip.csv')
allcf = pd.read_csv('CSV_DFs/CF_All_Precip.csv')

#Centered 49-hour window - doubled#
high49 = pd.read_csv('CSV_DFs/CF_High_49hr.csv')
low49 = pd.read_csv('CSV_DFs/CF_Low_49hr.csv')
all49 = pd.read_csv('CSV_DFs/CF_All_49hr.csv')

#Centered 13-hour window - halved#
high13 = pd.read_csv('CSV_DFs/CF_High_13hr.csv')
low13 = pd.read_csv('CSV_DFs/CF_Low_13hr.csv')
all13 = pd.read_csv('CSV_DFs/CF_All_13hr.csv')

#Left shifted 25-hour window#
highL = pd.read_csv('CSV_DFs/CF_High_LShr.csv')
lowL = pd.read_csv('CSV_DFs/CF_Low_LShr.csv')
allL = pd.read_csv('CSV_DFs/CF_All_LShr.csv')

#Right shifted 25-hour window#
highR = pd.read_csv('CSV_DFs/CF_High_RShr.csv')
lowR = pd.read_csv('CSV_DFs/CF_Low_RShr.csv')
allR = pd.read_csv('CSV_DFs/CF_All_RShr.csv')
```

```python
###Compare the dates for similarity###

#Ensure the date columns are actual dates#
for df in (high, low, allcf, highR, lowR, allR):
    df['date'] = pd.to_datetime(df['date']).dt.date

#Compute intersections#
def compare_dates(df1, df2, label):
    
    set1 = set(df1['date'])
    set2 = set(df2['date'])
    
    common = set1 & set2
    perc = len(common) / len(df1['date']) * 100
    
    print(f"{label}: {len(common):3d} dates in common or {perc}% match")
    
    return common

print("Right shifting the precipiation window")
common_high = compare_dates(high,  highR,  "High-precip")
common_low  = compare_dates(low,   lowR,   "Low-precip")
common_all  = compare_dates(allcf, allR,   "All cases")
```
